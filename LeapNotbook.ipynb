```python
from leap_finger import Finger
from leap_thumb import Thumb
from leap_palm import Palm
from leap_hand import LeapHand

# PyMJCF
from dm_control import mjcf

import PIL.Image


from dm_control.locomotion.arenas import floors
```

```python

finger = Finger()
print(finger.model.to_xml_string())
```

```python
# Instantiate the physics and render.
physics = mjcf.Physics.from_mjcf_model(finger.model)
PIL.Image.fromarray(physics.render())
```
